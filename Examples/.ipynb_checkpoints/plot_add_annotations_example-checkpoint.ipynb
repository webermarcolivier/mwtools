```python
mpnAnnotFeaturesDf = import_last_annotation_df(mpnAnnotationPath)
mpnAnnotFeaturesDf.loc['CDS'].head(2)
```

```python
def plot_coverage(covDf, xmin, xmax, yCols=None, annotFeaturesDf=None, saveFilePath=None,
                  ymin=None, ymax=None, figsize=(10, 6)):
    
    
#     xticks_locator = matplotlib.ticker.MaxNLocator(nbins=xticks_nbins)

    if yCols is None:
        yCols = covDf.columns
        yCols = [col for col in yCols if col != 'position']
    
    plotDf = covDf
    if covDf.index.name == 'position':
        plotDf = plotDf.reset_index()
    plotDf = plotDf[(plotDf['position'] >= xmin) & (plotDf['position'] <= xmax)]

    fig, ax = plt.subplots(figsize=figsize)    
    plotDf.plot(ax=ax, x='position', y=yCols)
    ax.set_xlim([xmin, xmax])
    ax.set_yscale('log')
    if ymin is not None and ymax is not None:
        ax.set_ylim([ymin, ymax])
    
    if annotFeaturesDf is not None:
        plot_add_annotations(ax=ax, fig=fig, annotFeaturesDf=annotFeaturesDf, show_ncRNA=True,
                             show_transcript=True, show_tRNA=True, idCol='id')
#     plt.subplots_adjust(hspace=0.01, bottom=0.03)

    if saveFilePath is not None:
        plt.savefig(str(saveFilePath), dpi=300, bbox_inches = 'tight')
        plt.close()

    return fig
```

```python
mpnPlot = mpnAnnotFeaturesDf.loc['CDS'].loc['MPN325']

xPad = 500
xmin = mpnPlot['start'] - xPad
xmax = mpnPlot['end'] + xPad

plot_coverage(covDf, xmin, xmax, yCols=None, annotFeaturesDf=mpnAnnotFeaturesDf, saveFilePath=None,
              ymin=None, ymax=None, figsize=(10, 6));
```

```python
xmin = mpnPlot['start'] - 10
xmax = mpnPlot['start'] + 80

plotDf = covDf
if covDf.index.name == 'position':
    plotDf = plotDf.reset_index()
plotDf = plotDf[(plotDf['position'] >= xmin) & (plotDf['position'] <= xmax)]

fig, ax = plt.subplots(figsize=(10, 6))    
plotDf.plot(ax=ax, x='position', y=['coverage_p', 'coverage_m'])
ax.set_xlim([xmin, xmax])
ax.set_yscale('log')


# Add DNA sequence highlighted feature
DNAseqFontPropertiesFeatureList = [{'start':0, 'end':3, 'background_color':'red',
                                    'background_color_alpha':0.4, 'weight':900}]

plot_add_annotations(ax=ax, fig=fig, annotFeaturesDf=mpnAnnotFeaturesDf,
                     show_transcript=True, show_ncRNA=True,
                     DNAFeatList=[('MPN325', DNAseqFontPropertiesFeatureList)])
```
